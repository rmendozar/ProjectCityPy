```python
# Dependencies
import pandas as pd
import numpy as np

fb_df = pd.read_csv("Resources/fob_winter_2018.csv")
grouped = fb_df.groupby(["country","question","value"]).sum().reset_index()
grouped['% value'] = round(grouped['statistic'] / grouped['total_asked']*100,1)
print(len(grouped))
```

```python
questions = grouped.question.nunique()
countries = grouped.country.nunique()

print(f'there are {questions} different questions for {countries} countries')
```
